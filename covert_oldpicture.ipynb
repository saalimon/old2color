```
from sklearn.model_selection import train_test_split
SEED = 44000
```

```
import cv2
```

```
data = []
label = []
```

```
type(data)
```

```
data_in = os.listdir("./habr_demo/input")
```

```
label_in = os.listdir("./habr_demo/colorized")
```

```
for i in data_in:
  img = cv2.imread("./habr_demo/input/"+i)
  data.append(img)
```

```
for i in label_in:
  img = cv2.imread("./habr_demo/colorized/"+i)
  label.append(img)
```

```
train_data, test_data, train_label, test_label = train_test_split(data, label, test_size=0.2, random_state=SEED)
```

```
sample = random.choice(train_data)
plt.imshow(sample)
```

```
sample = random.choice(train_label)
image = cv2.cvtColor(sample, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```
